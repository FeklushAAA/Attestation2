```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
#1 датасет
def dataset_1 ():
    #количесвто элементов в каждой категории
    X_size_1 = np.random.randint(50,100)
    X_size_0 = np.random.randint(50,100)
    #создание случайного датасета и разделение на train set / test set (80/20)
    X_1 = np.random.vonmises(0.0, 1, size = (X_size_1,2)) * 2       #центральный круг - 1
    X_0 = np.random.vonmises(0.0, 1, size = (X_size_0,2)) * 4       #внешний круг - 0
    X = np.concatenate([X_1, X_0])
    Y_1 = np.ones(X_size_1)         #центральный круг - 1 
    Y_0 = np.zeros(X_size_0)        #внешний круг - 0        
    Y = np.concatenate([Y_1, Y_0])
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2)
    #KNN
    k_range = range (1,9)
    scores = {}
    scores_list =[]
    for k in k_range:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, Y_train)
        Y_pred = knn.predict(X_test)
        scores [k] = metrics.accuracy_score(Y_test, Y_pred)
        scores_list.append(metrics.accuracy_score(Y_test, Y_pred))

    plt.plot(k_range, scores_list)
    plt.ylabel('Точность')
    plt.xlabel('Коэфициент К')
    
    mse = mean_squared_error(Y_test, Y_pred)      #корень среднеквадратической ошибки
    mae = mean_absolute_error(Y_test, Y_pred)     #абсолютная ошибка
    accuracy = metrics.accuracy_score(Y_test, Y_pred)
    f1_1 = metrics.f1_score(Y_test, Y_pred)
    print (f"F1 = {f1_1 /1:.4f}")
    print(f"Корень среднеквадратической ошибки = {mse /1:.4f}")
    print(f"Абсолютная ошибка = {mae /1:.4f}")
    print(f"Точность = {accuracy /1:.4f}")

    average_1 = metrics.average_precision_score(Y_test, Y_pred)
    print (f"Средняя точность = {average_1}")
    
dataset_1()
```

```python
C="Cat"
F="Fish"
H="Hen"
# истинные значения
y_true = [C,C,C,C,C,C, F,F,F,F,F,F,F,F,F,F, H,H,H,H,H,H,H,H,H]
# результат системы
y_pred = [C,C,C,C,H,F, C,C,C,C,C,C,H,H,F,F, C,C,C,H,H,H,H,H,H]
#расчёты
array_count = len([item for item in y_true])
def Cat ():
    С_TP = 0
    С_FN = 0
    С_FP = 0
    С_TN = 0
    C_support = 0
    number_c = 0
    #подсчёт TP, FN, FP, TN
    while number_c < array_count:
        if y_true[number_c] =="Cat" and y_pred[number_c] =="Cat":
            С_TP += 1
        elif y_true[number_c] =="Cat" and y_pred[number_c] !="Cat":
            С_FN += 1
        elif y_true[number_c] !="Cat" and y_pred[number_c] =="Cat":
            С_FP += 1
        elif y_true[number_c] !="Cat" and y_pred[number_c] !="Cat":
            С_TN += 1     
        number_c += 1
    #подсчёт количества Cat    
    number_c_1 = 0
    while number_c_1 < array_count:
        if y_true[number_c_1] =="Cat":
            C_support += 1
        number_c_1 += 1
    #метрики    
    С_precision = С_TP/(С_TP + С_FP)
    С_recal = С_TP/(С_TP + С_FN)
    С_f1 = (1 + np.square(1)) * (С_precision * С_recal)/(np.square(1) * С_precision + С_recal)
    #print(f"Cat        {С_precision/ 1:.3f}         {С_recal/ 1:.3f}         {С_f1/ 1:.3f}            {C_support}") 
    return C_support, С_precision, С_recal, С_f1
def Fish ():
    F_TP = 0
    F_FN = 0
    F_FP = 0
    F_TN = 0
    F_support = 0
    number_f = 0
    #подсчёт TP, FN, FP, TN
    while number_f < array_count:
        if y_true[number_f] =="Fish" and y_pred[number_f] =="Fish":
            F_TP += 1
        elif y_true[number_f] =="Fish" and y_pred[number_f] !="Fish":
            F_FN += 1
        elif y_true[number_f] !="Fish" and y_pred[number_f] =="Fish":
            F_FP += 1
        elif y_true[number_f] !="Fish" and y_pred[number_f] !="Fish":
            F_TN += 1     
        number_f += 1
    #подсчёт количества Cat    
    number_f_1 = 0
    while number_f_1 < array_count:
        if y_true[number_f_1] =="Fish":
            F_support += 1
        number_f_1 += 1
    #метрики    
    F_precision = F_TP/(F_TP + F_FP)
    F_recal = F_TP/(F_TP + F_FN)
    F_f1 = (1 + np.square(1)) * (F_precision * F_recal)/(np.square(1) * F_precision + F_recal)
    #print(f"{F}       {F_precision/ 1:.3f}         {F_recal/ 1:.3f}         {F_f1/ 1:.3f}           {F_support}")
    return F_support, F_precision, F_recal, F_f1
def Hen ():
    H_TP = 0
    H_FN = 0
    H_FP = 0
    H_TN = 0
    H_support = 0
    number_h = 0
    #подсчёт TP, FN, FP, TN
    while number_h < array_count:
        if y_true[number_h] =="Hen" and y_pred[number_h] =="Hen":
            H_TP += 1
        elif y_true[number_h] =="Hen" and y_pred[number_h] !="Hen":
            H_FN += 1
        elif y_true[number_h] !="Hen" and y_pred[number_h] =="Hen":
            H_FP += 1
        elif y_true[number_h] !="Hen" and y_pred[number_h] !="Hen":
            H_TN += 1     
        number_h += 1
    #подсчёт количества Cat    
    number_h_1 = 0
    while number_h_1 < array_count:
        if y_true[number_h_1] =="Hen":
            H_support += 1
        number_h_1 += 1
    #метрики    
    H_precision = H_TP/(H_TP + H_FP)
    H_recal = H_TP/(H_TP + H_FN)
    H_f1 = (1 + np.square(1)) * (H_precision * H_recal)/(np.square(1) * H_precision + H_recal)
    #print(f"{H}        {H_precision/ 1:.3f}         {H_recal/ 1:.3f}         {H_f1/ 1:.3f}            {H_support}")
    return H_support, H_precision, H_recal, H_f1

#расчёты
Cat()
C_support, С_precision, С_recal, С_f1 = Cat()
Fish()
F_support, F_precision, F_recal, F_f1 = Fish()
Hen()
H_support, H_precision, H_recal, H_f1 = Hen()

C_weighted = C_support/array_count
F_weighted = F_support/array_count
H_weighted = H_support/array_count

number = 0
cprrect_pred = 0
while number < array_count:
        if y_true[number] == y_pred[number]:
            cprrect_pred += 1
        number += 1
accuaracy = cprrect_pred/array_count

macro_avg_precision = 0.3333 * (С_precision + F_precision + H_precision)
macro_avg_recal = 0.3333 * (С_recal + F_recal+ H_recal)
macro_avg_f1 = 0.3333 * (С_f1 + F_f1+ H_f1)

weighted_avg_precision = C_weighted * С_precision + F_weighted * F_precision + H_weighted * H_precision
weighted_avg_recal = C_weighted * С_recal + F_weighted * F_recal + H_weighted * H_recal
weighted_avg_f1 =C_weighted * С_f1 + F_weighted * F_f1 + H_weighted * H_f1

#вывод данных
print("               precision        recall      f1_score      support")
print()
print(f"        {C}        {С_precision/ 1:.3f}         {С_recal/ 1:.3f}         {С_f1/ 1:.3f}            {C_support}")
print(f"        {F}       {F_precision/ 1:.3f}         {F_recal/ 1:.3f}         {F_f1/ 1:.3f}           {F_support}")
print(f"        {H}        {H_precision/ 1:.3f}         {H_recal/ 1:.3f}         {H_f1/ 1:.3f}            {H_support}")
print()
print(f"accuracy                                       {accuaracy/ 1:.3f}")  
print(f"macro avg          {macro_avg_precision/ 1:.3f}         {macro_avg_recal/ 1:.3f}         {macro_avg_f1/ 1:.3f}           {array_count}")
print(f"weighted avg       {weighted_avg_precision/ 1:.3f}         {weighted_avg_recal/ 1:.3f}         {weighted_avg_f1/ 1:.3f}           {array_count}")
```
